```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("./dz_files/dollar.xlsx")
```

```python
print(df)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['curs'])
plt.title('Динамика курса рубля')
plt.xlabel('Дата')
plt.ylabel('Курс рубля к доллару')
plt.grid(True)
plt.show()
```

```python
df1 = pd.read_csv("./dz_files/infl.csv")
```

```python
print(df1.columns.tolist())
print(df1)
```

```python
df1['Mean'] = df1[['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']].mean(axis=1)
```

```python
df1 = df1[df1['Year'] >= 2018]
```

```python
print(df1)
```

```python
df1['Mean'] = df1['Mean'].div(100)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df1['Year'], df1['Mean'])
plt.title('Динамика инфляции рубля')
plt.xlabel('Дата')
plt.ylabel('Динамика инфляции')
plt.grid(True)
plt.show()
```
